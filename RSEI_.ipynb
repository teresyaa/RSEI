## **Mengimport *Google Earth Engine***

## **Proses Indikator *Enhanced Vegetation Index* (EVI) dan representasi hasil EVI** 

````python
##ADMINISTRASI INDONESIA
country = ee.FeatureCollection("FAO/GAUL/2015/level0").filter(ee.Filter.eq('ADM0_NAME', 'Indonesia'))
````

````python
evi_2001 = ee.ImageCollection("MODIS/006/MOD13A2").select('EVI').filterDate('2001-01-01','2001-12-31').mean().divide(1e4).clip(country)

###################### EVI 2001 #####################
evi01_min = ee.Number(evi_2001.reduceRegion(**{
  'reducer': ee.Reducer.min(),
  'geometry': country,
  'scale': 1000,
  'maxPixels': 1e9}).values().get(0));

evi01_max = ee.Number(evi_2001.reduceRegion(**{
  'reducer': ee.Reducer.max(),
  'geometry': country,
  'scale': 1000,
  'maxPixels': 1e9}).values().get(0));

evi01_minmax =((evi_2001.subtract(evi01_min)).divide(evi01_max.subtract(evi01_min)))

eviVis = { 
    'min': -0.2,
    'max': 1,
    'palette': ['040274', '040281', '0502a3', '0502b8', '0502ce', '0502e6',
                '0602ff', '235cb1', '307ef3', '269db1', '30c8e2', '32d3ef',
                '3be285', '3ff38f', '86e26f', '3ae237', 'b5e22e', 'd6e21f',
                'fff705', 'ffd611', 'ffb613', 'ff8b13', 'ff6e08', 'ff500d',
                'ff0000', 'de0101', 'c21301', 'a71001', '911003']
    }

Map = geemap.Map(center = [-4.524586192827046,121.11328125], zoom = 5)
Map.addLayer(evi_2001, eviVis, 'EVI_2001')
Map
````

## **Proses Indikator *Land Surface Temperature* (LST) dan representasi hasil LST** 

````python
# Import the MODIS land surface temperature collection.
lst_2001 = ee.ImageCollection('MODIS/006/MOD11A2').select('LST_Day_1km') \
            .filterDate('2001-01-01','2001-12-31').mean().clip(country).multiply(0.02).subtract(273.15)

# Import the MODIS land surface temperature 2001 collection.
lst01_min = ee.Number(lst_2001.reduceRegion(**{
  'reducer': ee.Reducer.min(),
  'geometry': country,
  'scale': 1000,
  'maxPixels': 1e9}).values().get(0));

lst01_max = ee.Number(lst_2001.reduceRegion(**{
  'reducer': ee.Reducer.max(),
  'geometry': country,
  'scale': 1000,
  'maxPixels': 1e9}).values().get(0));

lst01_minmax =((lst_2001.subtract(lst01_min)).divide(lst01_max.subtract(lst01_min)))

lstVis = { 
    'min': 20,
    'max': 40,
    'palette': ['040274', '040281', '0502a3', '0502b8', '0502ce', '0502e6',
                '0602ff', '235cb1', '307ef3', '269db1', '30c8e2', '32d3ef',
                '3be285', '3ff38f', '86e26f', '3ae237', 'b5e22e', 'd6e21f',
                'fff705', 'ffd611', 'ffb613', 'ff8b13', 'ff6e08', 'ff500d',
                'ff0000', 'de0101', 'c21301', 'a71001', '911003']
    }

Map.addLayer(lst_2001, lstVis, 'LST_2001')
Map
````

## **Proses Indikator *Wetness* dan representasi hasil** 

````python
image01 = ee.ImageCollection("MODIS/006/MOD09GA").filterDate('2001-01-01','2001-12-31').mean().divide(1e4)
color_01 = image01.expression(
        '(0.1147 * B1) + (0.2489 * B2) + (0.2408 * B3) + (0.3132 * B4) + (0.3122 * B5) + (0.6416 * B6) + (0.5087 * B7)', {
        'B1' : image01.select('sur_refl_b01'),
        'B2' : image01.select('sur_refl_b02'),
        'B3' : image01.select('sur_refl_b03'),
        'B4' : image01.select('sur_refl_b04'),
        'B5' : image01.select('sur_refl_b05'),
        'B6' : image01.select('sur_refl_b06'),
        'B7' : image01.select('sur_refl_b07')})
wet_reduce01 = color_01.reduce(ee.Reducer.mean())
wet_2001 = wet_reduce01.clip(country)
###################### WET 2001 #####################
wet01_min = ee.Number(wet_2001.reduceRegion(**{
  'reducer': ee.Reducer.min(),
  'geometry': country,
  'scale': 1000,
  'maxPixels': 1e9}).values().get(0));

wet01_max = ee.Number(wet_2001.reduceRegion(**{
  'reducer': ee.Reducer.max(),
  'geometry': country,
  'scale': 1000,
  'maxPixels': 1e9}).values().get(0));

wet01_minmax =((wet_2001.subtract(wet01_min)).divide(wet01_max.subtract(wet01_min)))

wetVis = { 
    'min': -0.2,
    'max': 1,
    'palette': ['040274', '040281', '0502a3', '0502b8', '0502ce', '0502e6',
                '0602ff', '235cb1', '307ef3', '269db1', '30c8e2', '32d3ef',
                '3be285', '3ff38f', '86e26f', '3ae237', 'b5e22e', 'd6e21f',
                'fff705', 'ffd611', 'ffb613', 'ff8b13', 'ff6e08', 'ff500d',
                'ff0000', 'de0101', 'c21301', 'a71001', '911003']
    }

Map.addLayer(wet_2001, wetVis, 'WET_2001')
Map
````

## **Proses Indikator *Normalized Differential Build-Up and Bare Soil Index* (NDBSI) dan representasi hasil NDBSI** 

````python
IBI_01 = image01.expression(
        '(((2*B6)/(B6+B2))-(((B2)/(B2+B1))+((B4)/(B4+B6))))/(((2*B6)/(B6+B2))+(((B2)/(B2+B1))+((B4)/(B4+B6))))', {
        'B1' : image01.select('sur_refl_b01'),
        'B2' : image01.select('sur_refl_b02'),
        'B3' : image01.select('sur_refl_b03'),
        'B4' : image01.select('sur_refl_b04'),
        'B5' : image01.select('sur_refl_b05'),
        'B6' : image01.select('sur_refl_b06'),
        'B7' : image01.select('sur_refl_b07')})

BI_01 = image01.expression(
        '((B6+B1)-(B2+B3))/((B6+B1)+(B2+B3))', {
        'B1' : image01.select('sur_refl_b01'),
        'B2' : image01.select('sur_refl_b02'),
        'B3' : image01.select('sur_refl_b03'),
        'B4' : image01.select('sur_refl_b04'),
        'B5' : image01.select('sur_refl_b05'),
        'B6' : image01.select('sur_refl_b06'),
        'B7' : image01.select('sur_refl_b07')})

colorized01 = ((IBI_01).add(BI_01)).divide(2)
dry_reduce01=colorized01.reduce(ee.Reducer.mean())
dry_2001 = dry_reduce01.clip(country)

###################### DRY 2001 #####################
dry01_min = ee.Number(dry_2001.reduceRegion(**{
  'reducer': ee.Reducer.min(),
  'geometry': country,
  'scale': 1000,
  'maxPixels': 1e9}).values().get(0));

dry01_max = ee.Number(dry_2001.reduceRegion(**{
  'reducer': ee.Reducer.max(),
  'geometry': country,
  'scale': 1000,
  'maxPixels': 1e9}).values().get(0));

dry01_minmax =((dry_2001.subtract(dry01_min)).divide(dry01_max.subtract(dry01_min)))

dryVis = { 
    'min': -0.2,
    'max': 1,
    'palette': ['040274', '040281', '0502a3', '0502b8', '0502ce', '0502e6',
                '0602ff', '235cb1', '307ef3', '269db1', '30c8e2', '32d3ef',
                '3be285', '3ff38f', '86e26f', '3ae237', 'b5e22e', 'd6e21f',
                'fff705', 'ffd611', 'ffb613', 'ff8b13', 'ff6e08', 'ff500d',
                'ff0000', 'de0101', 'c21301', 'a71001', '911003']
    }
Map.addLayer(dry_2001, dryVis, 'DRY_2001')
Map
````

## **Menggabungkan 4 indikator dalam 1 layer**

````python
image_01 = ee.Image([lst01_minmax, evi01_minmax, dry01_minmax, wet01_minmax]).toDouble()
````

## **Proses *Principal Component Analysis* (PCA)** 

````python
# Set some information about the input to be used later.
scale = 1000
bandNames = image_01.bandNames()
region = country

# Mean center the data to enable a faster covariance reducer
# and an SD stretch of the principal components.
meanDict = image_01.reduceRegion(**{
    'reducer': ee.Reducer.mean(),
    'geometry': region,
    'scale': scale,
    'maxPixels': 1e9
})
means = ee.Image.constant(meanDict.values(bandNames))
centered = image_01.subtract(means)

# This helper function returns a list of new band names.
def getNewBandNames(prefix):
  seq = ee.List.sequence(1, bandNames.length())
  return seq.map(lambda b: ee.String(prefix).cat(ee.Number(b).int().format()))


# This function accepts mean centered imagery, a scale and
# a region in which to perform the analysis.  It returns the
# Principal Components (PC) in the region as a new image.
def getPrincipalComponents(centered, scale, region):
  # Collapse the bands of the image into a 1D array per pixel.
  arrays = centered.toArray()

  # Compute the covariance of the bands within the region.
  covar= arrays.reduceRegion(**{
    'reducer': ee.Reducer.centeredCovariance(),
    'geometry': region,
    'scale': scale,
    'maxPixels': 1e9
  })

  # Get the 'array' covariance result and cast to an array.
  # This represents the band-to-band covariance within the region.
  covarArray = ee.Array(covar.get('array'))

  # Perform an eigen analysis and slice apart the values and vectors.
  eigens = covarArray.eigen()

  # This is a P-length vector of Eigenvalues.
  eigenValues = eigens.slice(1, 0, 1)
  # This is a PxP matrix with eigenvectors in rows.
  eigenVectors = eigens.slice(1, 1)

  # Convert the array image to 2D arrays for matrix computations.
  arrayImage = arrays.toArray(1)

  # Left multiply the image array by the matrix of eigenvectors.
  principalComponents = ee.Image(eigenVectors).matrixMultiply(arrayImage)

  # Turn the square roots of the Eigenvalues into a P-band image.
  sdImage = ee.Image(eigenValues.sqrt()) \
    .arrayProject([0]).arrayFlatten([getNewBandNames('sd')])

  # Turn the PCs into a P-band image, normalized by SD.
  return principalComponents \
    .arrayProject([0]) \
    .arrayFlatten([getNewBandNames('pc')]) \
    .divide(sdImage) \


# Get the PCs at the specified scale and in the specified region
pcImage01 = getPrincipalComponents(centered, scale, region)
````

#**Klasifikasi RSEI**

````python
palette = ['ff1700', 'ff8e00', 'ffe400', '06ff00', '2cba00' ]
````

````python
rsei_min01 = ee.Number(pcImage01.select(0).reduceRegion(**{
  'reducer': ee.Reducer.min(),
  'geometry': country,
  'scale': 1000,
  'maxPixels': 1e9}).values().get(0));

rsei_max01 = ee.Number(pcImage01.select(0).reduceRegion(**{
  'reducer': ee.Reducer.max(),
  'geometry': country,
  'scale': 1000,
  'maxPixels': 1e9}).values().get(0));

rsei01 =((pcImage01.select(0).subtract(rsei_min01)).divide(rsei_max01.subtract(rsei_min01)))
````

## **Representasi Hasil**

````python
Map = geemap.Map(center = [-4.524586192827046,121.11328125], zoom = 5)
Map.addLayer(rsei01, {'palette': palette, 'min' : 0, 'max': 1}, 'RSEI 2001')
Map
````
